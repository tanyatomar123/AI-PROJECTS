```python
import re
import random

class RuleBot:
    negative_responses = ("no", "nope", "nah", "not a chance", "sorry")
    exit_commands = ("quit", "pause", "exit", "goodbye", "bye", "later")
    random_questions = (
        "Why are you here?",
        "Are you studying?",
        "Do you have friends?",
        "Do you keep things straight?",
        "What kind of friends do you want?",
        "What is your plan today?",
    )

    def __init__(self):
        self.friendBabble = {
            'describe_plan_intent': r'.*\s*your plan.*',
            'answer_why_intent': r'why\sare.*',
            'about_friends': r'.*\s*friends'
        }

    def greet(self):
        self.name = input("What is your name?\n")
        will_help = input(
            f"Hi {self.name}, I am a chatbot. Will you help me know about you?\n"
        )
        if will_help.lower() in self.negative_responses:
            print("Okay, have a nice day!")
            return
        self.chat()

    def make_exit(self, reply):
        for command in self.exit_commands:
            if reply == command:
                print("Okay, have a nice day!")
                return True
        return False

    def chat(self):
        reply = input(random.choice(self.random_questions)).lower()
        while not self.make_exit(reply):
            reply = input(self.match_reply(reply))

    def match_reply(self, reply):
        for key, value in self.friendBabble.items():
            intent = key
            regex_pattern = value
            found_match = re.match(regex_pattern, reply)
            if found_match and intent == 'describe_plan_intent':
                return self.describe_plan_intent()
            elif found_match and intent == 'answer_why_intent':
                return self.answer_why_intent()
            elif found_match and intent == 'about_friends':
                return self.about_friends()
        return self.no_match_intent()

    def describe_plan_intent(self):
        responses = (
            "My plan is to study and hang out with friends.\n",
            "It has been a while.\n"
        )
        return random.choice(responses)

    def answer_why_intent(self):
        responses = (
            "I am here to talk.\n",
            "I am here to share thoughts.\n",
            "I heard you need friends.\n"
        )
        return random.choice(responses)

    def about_friends(self):
        responses = (
            "Friends are the sweetest creatures created by God.\n",
            "Friends are always with us, even when you feel low.\n",
            "The heart of a friend is always pure.\n"
        )
        return random.choice(responses)

    def no_match_intent(self):
        responses = (
            "Please tell me more.\n", 
            "Tell me more!\n",
            "I see. Can you explain?\n",
            "Interesting. Can you tell me more?\n",
            "Why?\n", 
            "How do you think I feel when you say this?\n"
        )
        return random.choice(responses)

friendBot = RuleBot()
friendBot.greet()
```
